```python
import matplotlib.pyplot as plt
import numpy as np
#Line Plot
x=[1,2,3,4]
y=[11,22,33,44]

plt.plot(x,y,color='red',linestyle='-',marker='o',label='Line Plot')
plt.legend()
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#scatter plot
```

```python

#scatter plot
plt.scatter(x,y,color='blue',marker='*',label='Scatter Plot')
plt.legend()
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python


# Bar plot
plt.barh(x, y, color='green', label='Bar Plot')
plt.legend()
plt.title('Bar Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# prompt: histogram

import matplotlib.pyplot as plt
import numpy as np

# generate some random data for the histogram
data = np.random.randn(1000)

plt.hist(data, bins=30, color='red', alpha=0.7, label='Histogram')
plt.legend()
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# prompt: pie

import matplotlib.pyplot as plt
import numpy as np

# Pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['blue', 'green', 'red', 'yellow']
explode = (0, 0.1, 0, 0)  # explode the 2nd slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart')
plt.show()
```

```python
# prompt: box plot

# generate some random data for the box plot
data = np.random.randn(100, 5)

plt.boxplot(data, labels=['A', 'B', 'C', 'D', 'E'], patch_artist=True, vert=True)

plt.title('Box Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# prompt: violin plot

import matplotlib.pyplot as plt
import numpy as np

# generate some random data for the violin plot
data = [2,3,4,5,6,7,8,9],[1,6,7,8,9],[4,6,7,8]

plt.violinplot(data, showmeans=True, showmedians=True)

plt.title('Violin Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```
