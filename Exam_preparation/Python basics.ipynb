```python
#read a character
#v-violet,i=indigo,b=blue,g=green,y=yellow,o=orange,r=red,any other character invalid color

char=input("Enter a Character")
if char=="v" or char=="V":
    print("violet")
elif char=="i" or char=="I":
    print("indigo")
elif char=="b" or char=="B":
    print("blue")
elif char=="g" or char=="G":
    print("green")
elif char=="y" or char=="Y":
    print("yellow")
elif char=="o" or char=="O":
    print("orange")
elif char=="r" or char=="R":
    print("red")
else:
    print("invalid character")
```

```python
#print odd no.s from list

numList=[12,23,54,19,76]

count=0
j=0
while j<len(numList):
    if numList[j]%2==1:
        count=count+1
    j=j+1
print("Total odd numbers: ",count)
    
```

```python
#max element

max=0
j=0

while j<len(numList):
    if numList[j]>max:
        max=numList[j]
    j=j+1
print("The max ele")
```
